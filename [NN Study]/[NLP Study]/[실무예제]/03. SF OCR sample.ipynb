```python
import os

url = 'UseData/SmartFarmOCR/'
os.listdir(url)
```

```python
file = url + os.listdir(url)[0]
file
```

```python
with open(file, 'r', encoding='utf-8-sig') as f:
    myList = []
    # f.read().replace('\n', '') 하면 1글자씩 찢어짐.
    # f.readline() 하면 공백기준 첫글자만.
    # f.readlines() 하면 전체 읽어서 list로 반환.
    lines = f.readlines()
    for line in lines:
        myList.append(line.replace('\n', ''))
```

```python
for whitespace in myList:
    if whitespace == '' or whitespace == ' ':
        myList.remove(whitespace)
#for i in range(len(myList)):
#    if myList[i] == '' or myList[i] == ' ':
#        myList.remove(myList.index(i))
```
