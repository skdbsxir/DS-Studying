```python
url = 'UseData/Voucher/Cleaned/'
os.listdir(url)
```

```python
filesDataFrame = pd.DataFrame({'File' : os.listdir(url)})
filesDataFrame
```

```python
sampleFile = filesDataFrame['File'][0]
sampleFile

df = pd.read_csv(url + sampleFile)
df
```

```python
def PosNegAdder(url, file):
    df = pd.read_csv(url + file)
    dropList = ['Unnamed: 0', 'Unnamed: 0.1']
    df.drop(dropList, axis='columns', inplace=True)
    df['긍부정'] = df['별점'].apply(lambda x: 1 if x>3 else 0)
    df.to_csv(url + file[:-4] + '.csv', index=True, header=True, encoding='utf-8-sig')
```

```python
for index in range(len(filesDataFrame)):
    file = filesDataFrame['File'][index]
    PosNegAdder(url, file)
```
