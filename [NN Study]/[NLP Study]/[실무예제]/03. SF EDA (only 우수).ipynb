```python
df[colList].isnull().sum()
```

```python
dropList = []
for variable in colList:
    # 95% 이상이 결측인 컬럼?
    if df[variable].isnull().sum() >= (0.95 * len(df[variable])):
        # print(variable)
        dropList.append(variable)
dropList
```

    해당 컬럼들 drop.

```python
for variable in colList:
    print(f'{variable} 속성 전체 데이터 수 : ', len(df[variable]))
    print(f'{variable} 속성 결측치 수 : ', df[variable].isnull().sum())
    print('\n')
```
