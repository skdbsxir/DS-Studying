```python
import os, warnings
import pandas as pd
from konlpy.tag import Okt

warnings.filterwarnings('ignore')
```

```python
url = 'UseData/Voucher/Cleaned/'
os.listdir(url)
```

     일단 한 파일만 불러와서 Okt 써서 컨텐츠 분리?분석?을 해보자.

```python
samplefile = os.listdir(url)[0]
samplefile
```

```python
df = pd.read_csv(url + samplefile)
dropList = ['Unnamed: 0', '제목']

df.drop(dropList, axis='columns', inplace=True)
df
```

```python
okt = Okt()
df['Contents_norm'] = df['Contents'].apply(lambda x: okt.normalize(x))
df
```

```python
df['Contents_morphs'] = df['Contents_norm'].apply(lambda x: okt.morphs(x))
df
```

```python
df['Contents_noun'] = df['Contents_norm'].apply(lambda x: okt.nouns(x))
df
```

```python
df['Contents_phrase'] = df['Contents_norm'].apply(lambda x: okt.phrases(x))
df
```
