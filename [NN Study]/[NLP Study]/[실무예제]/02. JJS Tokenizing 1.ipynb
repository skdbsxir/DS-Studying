```python
okt = Okt()
df['Contents_pos'] = df['Contents'].apply(lambda x: okt.pos(x, norm=True)) # stem을 하는게 좋을까 흠
df
```

```python
df['Contents_pos2'] = df['Contents'].apply(lambda x: okt.pos(x, norm=True, stem=True))
df
```

```python
df['Contents_pos2'][3]
```

```python
import nltk
import numpy as np

res = []
uni = {}

for sentence in df['Contents']:
    temp_x = okt.pos(sentence)
    if all(np.in1d(list(set([x[1] for x in temp_x])), ['Alpha', 'Number'])) :
        words = nltk.tokenize.word_tokenize(sentence)
        temp_x = nltk.tag.pos_tag(words)
        temp_x = [word for word, morp in temp_x if ('NN' in morp)]
        for t in temp_x :
            if t in uni.keys() :
                uni[t] += 1
            else :
                uni[t] = 1
        res.append(temp_x)
        continue
    temp_x = [word for word, morp in temp_x if (len(word) > 1) & (morp in ['Noun', 'Alpha', 'Number'])]
    for t in temp_x :
        if t in uni.keys() :
            uni[t] += 1
        else :
            uni[t] = 1
    res.append(temp_x)
```
