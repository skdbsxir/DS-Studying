## R 외부 파일 읽기

<br>

- scan() 함수 : 외부파일 읽기.
    - **scan('파일명', what)**
        - what : 문자열인 경우 "" or 'char'이라고 입력해줘야만 인식 가능. (what = 'char')
        - 추가 인자로 sep 을 지정해 무얼 기준으로 잘라서 읽을지 지정 가능.
            - (python read_csv의 sep 인자랑 같은 기능.)
    - 크지 않은 파일을 읽어올 때 유용.
    - 사용자 입력을 받을때도 사용. 엔터를 두번치면 입력 종료. (R console에서는 되는듯?)
        - 사용자 입력은 주로 readLine 사용. 이전에 썼던 것 처럼.
        

```r
a <- scan('UseData/irum.txt', what = 'char')
a

# sep 인자를 통해 ,를 구분으로 해서 읽어옴.
b <- scan('UseData/irum.txt', what = 'char', sep = ',')
b
```

- readLines() : 한 줄 단위로 읽어옴.
    - **readLines('파일 명')**
    * 주의 : readLine()은 줄 단위로 데이터를 입력하는 것.

```r
data <- readLines('UseData/consumer.txt')
data
```

- read.csv() : csv 파일 읽어옴.
    - **read.csv('파일 명')**
        - xlsx 파일은 읽어오는데 시간이 너무 오래걸리고, 용량이 너무 큼.
        - 그래서 보통 csv파일로 변환한 다음 csv파일로 읽어옴.
    - 마찬가지로 header=TRUE를 통해 첫 행을 column으로 지정 할 수 있음.

```r
data2 <- read.csv('UseData/seoulpopulation.csv', header=TRUE)
data2
```

**엑셀 파일 읽기?**

<br>

- **install.packages('XLConnect')**로 패키지 설치 후, **require(XLConnect)**를 통해 import.
- import한 후 **loadWorkbook('파일경로')**를 통해 불러올 수 있음.
- 그 후 readWorksheet() 사용해 DataFrame 형태로 읽어옴. 사용할 수 있는 인자는
    - sheet = '시트 명' (시트가 여러개 있는 경우)
    - startRow = 읽어올 시작 행
    - startCol = 읽어올 시작 열
    - endRow = 마지막 행
    - endCol = 마지막 열
- 데이터가 많아지면 속도가 좀 많이 느림.

```r
install.packages('XLConnect')
require(XLConnect)
data <- loadWorkbook('UseData/subway.xlsx') # 이게 왜없어 ㅋㅋㅋ
df <- readWorksheet(data, sheet=1, header=TRUE)
df
```

- XLConnect 말고 **xlsx패키지** 많이 씀.
- java를 이용해 xls, xlsx를 불러올 수 있음.
- **install.packages('xlsx')** 후에 **require(xlsx)**
- import 한 후 read.xlsx('파일경로') or **read.xlsx2('파일경로')**
    - read.xlsx2를 쓰면 한글이 깨지거나 하는 문제가 없음.
    - 파일경로 뒤에 sheet number를 지정해 특정 시트만 불러옴.

```r
install.packages('xlsx')
require(xlsx)
df <- read.xlsx2('UseData/subway.xlsx', 1)
df
```
