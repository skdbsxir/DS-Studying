### ggplot()

<br>

- **ggplot(data, aes(x = x축 데이터, y = y축데이터)) + geom_원하는그래프()**
    - geom은 원하는 그래프.
    - data : 그래픽월 원하는 데이터 집합
    - aes() : 데이터 각 레코드를 그래픽 요소에 매핑하는 작업.
    - geom() : 각 축에 숫자형 데이터를 입력받아 실제로 시각화 하기 위해 좌표계에 표시하는 작업.
        - 기하 객체. geometric object를 실제로 그리는 작업 수행.
        - 표시되는 형태에 따라 점(point), 선(line), 다각형(polygon) 등 많은 객체가 존재.
            - geom_line(), geom_bar(), geom_point() 처럼 사용.
    - stat : 해당 데이터 값에 대한 통계적 처리 방법.

```r
data1 <- data.frame(mat=c(55, 75, 80, 65, 90, 100, 70, 85), eng=c(65, 100, 45, 50, 75, 90, 90, 65),
                   avg=c(53, 70, 83, 70, 93, 95, 75, 80), 
                   name=c('김', '이', '박', '최', '문', '윤', '손', '정'))
data1

# x=~, y=~ 안하고 그냥 ~, ~해도 됨.
# +를 더 붙여서 추가로 더 그려줄 수 있음. 이를 '레이어 작업'이라 함.
 # 먼저 그린 그래프가 뒤로 상속되는 개념.
ggplot(data1, aes(x=avg, y=mat))+geom_line(col='blue')+geom_point(aes(color=name))
```

**레이어 상속?**

<br>

- geom_~() 로 그리면 먼저 그린게 뒤에꺼로 상속됨.
    - 먼저 그린것에 뒤에 그리는걸 덧씌우는 개념.

```r
ggplot(data1, aes(mat, avg))+geom_line(col='red')+geom_point()+
geom_line(aes(eng, avg), col='darkgreen')+geom_point(aes(color=name))
```

- 지금 보면 **eng line (녹색 선)에 point가 찍히지 않았음.** 
- mat line (빨간 선)에 흑색 점을 찍도록 했는데, eng line (녹색 선)에 그려줄 점이 mat line (빨간 선)에 그려짐.
    - 상속 되지 않은 것.
    - 상속 시키려면? eng line (geom_line(aes(eng, avg),~)에 넣은 인자들을 바로 뒤에 그대로 넣어줘야 함.

```r
ggplot(data1, aes(mat, avg))+geom_line(col='red')+geom_point()+
geom_line(aes(eng, avg), col='darkgreen')+geom_point(aes(eng, avg), color='darkgreen')
```

```r
datainfo <- read.csv('UseData/highwaybus09.csv', sep=',', header=TRUE)
ggplot(datainfo, aes(x=선별))+geom_bar()
ggplot(datainfo, aes(x=선별, y=총이용인원))+geom_bar(stat='identity', fill='orange', color='red')
```

    geom_bar() 에 stat요소는 기본적으로 stat_bin(). 빈도수 표시.
    처음 그린 그래프가 빈도수만 그린 그래프.
    항목 계산을 위해선 stat='identity'로 설정해줘야 함.

```r
library(ggplot2)
# 평균을 계산해서 시각화 한다면 미리 계산해줘야 함.
datainfo <- read.csv('UseData/highwaybus09.csv', sep=',', header=TRUE)
meandf <- as.data.frame(with(datainfo, tapply(이용율, 선별, mean, na.rm=TRUE)))
meandf$선 <- rownames(meandf)
names(meandf) <- c('이용율', '선')
meandf

ggplot(meandf, aes(선, 이용율))+geom_bar(stat='identity', aes(fill=선))
```

    영동선의 평균 이용률이 가장 높은 것을 확인할 수 있음.

**누적막대그래프**

```r
meandf1 <- as.data.frame(with(datainfo, aggregate(이용율, list(선별=선별, 차종=차종), mean, 
                                                  na.rm=TRUE)))
meandf1$이용율 <- rownames(meandf1)
names(meandf1) <- c('선별', '차종', '평균이용율')
meandf1

ggplot(meandf1, aes(선별, 평균이용율))+geom_bar(stat='identity', aes(fill=차종))
```

    평균 이용수가 많은 영동선은 고속의 이용이 많음을 알 수 있음.

```r
ggplot(meandf1, aes(선별, 평균이용율))+geom_bar(stat='identity', aes(fill=차종), position='fill')
```

    position='fill' 옵션을 추가하면 100%에 대한 누적 막대그래프가 나옴.

```r
ggplot(meandf1, aes(선별, 평균이용율))+geom_bar(stat='identity', aes(fill=차종), position='dodge')
```

    fill 대신 dodge를 넣으면 누적이 아닌, 2개의 label을 비교하는 식으로 나타남.

```r
# 시각화 된 결과에 계산한 평균값을 글자로 넣기?
ggplot(meandf1, aes(선별, 평균이용율))+geom_bar(stat='identity', aes(fill=차종), position='dodge')+
geom_text(aes(y=평균이용율, label=평균이용율), color='black', size=3)
```
