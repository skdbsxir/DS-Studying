## R Graphics

* plot() 함수 기본

    - **plot(x, y, 옵션)**
        - main : 그래프 제목
        - sub : 그래프 부제목
        - xlab, ylab : x, y축 제목
        - type : plot 형태
        - axes : plot의 테두리 선 (ex. axes = F : x, y축 테두리선 표시하지 않음.)
        - col : plot의 색상
        
    - plot.new()를 통해 기존 차트를 지우고 표시할 수 있음. (R Console에서 해당하는듯?)

```r
age <- c(23, 35, 43, 28, 55, 47, 58)
plot(age, main='연령분포', ylab='나이', col='red')
```

```r
height <- c(165, 170, 177, 167, 180, 172, 169, 179)
weight <- c(70, 73, 62, 80, 85, 74, 75, 68)
plot(weight, height, main='키와 몸무게의 분포', xlab='체중', ylab='키', col=rainbow(length(height)))
```

    벡터 x, y의 길이가 같으면 plot시 둘은 1:1 대응으로 그려짐.
    
    col = rainbow(길이) 지정을 통해 점들의 색을 임의배정 할 수 있음.

- type = '종류' 설정을 통해 그래프 형태 지정이 가능.
    - p(점), l(선), o(점과 선 중첩), h(히스토그램과 같이 수직선), s or S(계단형)

**par() 함수**

<br>

- **화면 분할** 효과. **여러 그래프를 한 화면에 그릴때** 유용.
    - python의 subplots같은 거.
- 선의 굵기와 종류, 문자의 크기와 글꼴, 색상 지정.
- par(인자) ***인자 종류가 많음.***
    - par(mfrow = c(행 개수, 열 개수))
        - par(mfrow = c(2, 3)) : 2행 3열의 그래프. 한 화면의 6개의 그래프가 그려짐.

```r
par(mfrow = c(1,3))
plot(c(1:3), main='a-type')
plot(c(2:5), c(2:5), main='b-type', type='c')
plot(1:5, main='c-type', type='l')
```

```r
# 좀 더 응용해서?
x <- c(1:5)
y <- c(1:5)
par(mfrow=c(2,3))
kind = c('p', 'l', 'o', 'b', 'S', 'h')

for(i in 1:length(kind)){
    title <- paste('type = ', kind[i])
    plot(x, y, type=kind[i], main=title, col=rainbow(length(kind)))
}
```

**points() 함수**

- 그래프에 나타나는 점 모양을 변경할 수 있음.
- points(x, y, pch = '점 형태(번호)', cex='점의 크기 기술(숫자)')

- chart를 돋보이게 하는 보조함수.
    - 따라서 사용한다면 plot함수 아래에 적어줌.

### pch, cex는 plot, par 함수 안에서도 사용되는 인자.

-----------
**lines() 함수**
- line(x, lty='선의 타입', col='선 색상', lwd='선의 굵기(숫자)')
    - dotted, solid, dashed, .... 타입 종류도 많음.

```r
plot(1:5, 1:5, type='n')
lines(c(1,3), c(3,3), lty='dotted', lwd=3, col='red')
lines(c(1,3), c(4,4), lty='solid', lwd=3, col='green')
lines(c(1,5), lty='dashed', lwd=4, col='blue')
```

## 그 외 보조 함수들?

<br>

- **abline()** : 그래프 안에 선 그리기 
    - abline(h=3, lty='solid')
- **text()** : 그래프 안에 텍스트 삽입
    - text(3, 2, '하위 data군')
- **legend(x, y, cex='글자크기', col='색상', lty='선종류')** : 그래프 안에 범례 삽입
- **axis(위치(1-아래, 2-왼쪽, 3-위, 4-오른쪽), at='그려지는 값', labels=그려지는 값의 범례, lty='선종류')**

```r
x <- c(165, 170, 173, 175, 180, 176, 172, 168)
y <- c(66, 70, 72, 80, 85, 78, 65, 62)

name <- c('a', 'b', 'c', 'd', 'e', 'f', 'g', 'h')

plot(x, y, type='b', pch=21, col=rainbow(length(x)))

abline(h=seq(65, 85, 5), col='gray', lty='dotted') # 65, 70, 75, 80, 85에 점선 그리기.
text(175, 73, '표준') # (175, 73) 위치에 글자 넣기

# 축 직접 설정. 2(왼쪽)에 at=x(x값들)을 그려줌.
# 근데 요즘엔 자동으로 그려줘서, 굳이 안넣어도 되긴 함.
axis(2, at=x, labels=x, col.axis='gray')  
legend('bottomright', col=rainbow(length(x)), legend=name, lty=1) # 우하단에 범례 추가.
```
