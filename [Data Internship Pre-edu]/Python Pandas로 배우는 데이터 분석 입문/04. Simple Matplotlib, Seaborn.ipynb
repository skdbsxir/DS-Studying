```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

"""
# 한글폰트 설정
plt.rcParams['font.family'] = 'NamumGothic'
plt.rcParams['axes.unicode_minus'] = False
"""
X_label = ['Ad1_ClickCount', 'Ad2_ClickCount', 'Ad3_ClickCount', 'Ad4_ClickCount', 'Ad5_ClickCount']
maleAdsClickCount = [20, 34, 30, 35, 27]
femaleAdsClickCount = [25, 32, 34, 20, 25]
```

## 이상하게 한글폰트가 자꾸 안나온다. </b>
    그냥 영어로 설정하는게 제일 나을듯.

```python
x = np.arange(len(X_label))
print(x)
```

```python
width = 0.35

# subplot 설정.
fig, ax = plt.subplots() 
fig.set_size_inches(12, 5)

# 각각 그래프 그리기
rects1 = ax.bar(x - width/2, maleAdsClickCount, width, label = 'Male')
rects2 = ax.bar(x + width/2, femaleAdsClickCount, width, label = 'Female')

ax.set_ylabel('Click Count')
ax.set_title('Ad Click Count by gender')
ax.set_xticks(x) # x축 데이터 설정
ax.set_xticklabels(X_label) # X축 레이블 설정
ax.legend() # 범례 설정

plt.show()
```

## * Seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
# Seaborn의 내장 데이터셋.
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.countplot(data=tips, x='day')
plt.title('Daily Tip Count ')
```

```python
sns.boxplot(data=tips, x='day', y='tip')
plt.title('Daily Tip Count - Boxplot')
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip')
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip')
```

```python
sns.boxplot(data=tips, x='day', y='tip', hue='sex')
```

```python
sns.violinplot(data=tips, x='day', y='tip', hue='sex')
```

```python
sns.distplot(tips.tip, rug=True)
# rug = True : show each observation with marginal ticks
```
