## Matplotlib
    2차원 그래프 작성에 주로 사용.
    OS의 제약이 X.
    그래프에 다양한 설정이 가능. 자세한 Handling이 가능.

```python
import matplotlib.pyplot as plt
```

- Figure : 그림을 그릴 수 있는 영역 자체. 도화지.
- Subplot : 실제 그림이 그려지는 영역. 도화지 안에서 그림그리는 영역.

<br>

## 그림 그리는 순서?
1. Figure 설정. 종이를 가져와서
2. Subplot 작업. 해당 영역에 그림을 그려라.

```python
fig = plt.figure() # 종이를 가져와라.
ax = fig.add_subplot(111) # 빈 종이에 그림 그릴 영역을 추가해라.
# add_subplot(xyz) ==> x : 행 번호, y : 열 번호, z : 그림 번호
data = [0,1]
ax.plot(data) # 데이터를 그려라.
plt.show() # 그린 그림을 화면에 출력해라.
```

```python
fig = plt.figure()

# 새 종이를 여러 등분으로 나눠보자.
ax1 = fig.add_subplot(221) # 2행 2열 1번째 그래프
ax2 = fig.add_subplot(222) # 2행 2열 2번째 그래프
ax3 = fig.add_subplot(223) # 2행 2열 3번쨰 그래프

plt.show()
```

    그래프 스타일을 변경할 수 있음.
    배경이라던지, 선 모양이라던지...

```python
# 배경에 바둑 격자무늬는 어떻게?
plt.style.use('ggplot') # 이렇게도 사용할 수 있음.
fig = plt.figure()
ax = fig.add_subplot(111)
data = [0, 1]
ax.plot(data)
plt.show()
```

```python
# 직선이 아닌 꺾은선 그래프는 어떻게?
fig = plt.figure()
ax = fig.add_subplot(111)
x = [0,2,4]
y = [0,4,2]
ax.plot(x, y)
plt.show()
```

**산포도 그래프**

<br>

    데이터의 추세, 패턴을 확인하는데 편함.
    plot()대신 scatter().

```python
import numpy as np
plt.style.use('ggplot')

# random을 이용해 데이터를 생성해보자.
np.random.seed(2)
x = np.arange(1,101)
y = 2 * x * np.random.rand(100)

fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(x, y)
plt.show()
```

**막대그래프**

<br>

    plot() 대신 bar()

```python
plt.style.use('ggplot')
fig = plt.figure()
ax = fig.add_subplot(111)
x = [3, 4]
y = [1, 2]

ax.bar(x, y)
plt.show()
```

## 그래프 레이블 지정

<br>

    plt.xlabel()을 통해 x축 레이블 지정.
    plt.ylabel()을 통해 y축 레이블 지정.
    plt.title()을 통해 그래프 제목 지정.
    bar() 안 인자로 tick_label을 통해 막대의 이름 지정.

```python
plt.style.use('ggplot')
fig = plt.figure()
ax = fig.add_subplot(111)
labels = ['english', 'math']
x = [3, 4]
y = [1, 2]

# tick_label인자로 label을 지정해주면 막대의 범례가 생김.
ax.bar(x, y, tick_label = labels)
plt.show()
```

```python
# 가로로 누은 그래프는 어떻게?
plt.style.use('ggplot')
fig = plt.figure()
ax = fig.add_subplot(111)
labels = ['english', 'math']
x = [3, 4]
y = [1, 2]

# bar가 아닌 barh 를 쓰면 됨.
ax.barh(x, y, tick_label=labels)
plt.show()
```

```python
# 여러 데이터를 묶어서 그리기?
x = [1, 2]
y1, y2, y3 = [1, 2], [2, 5], [3, 7]

fig = plt.figure()
ax = fig.add_subplot(111)

w = 0.2
ax.bar(x, y1, width = w, label = 'english')
ax.bar(np.array(x) + w, y2, width = w, label = 'math')
ax.bar(np.array(x) + w * 2, y3, width = w, label = 'korean')

ax.legend()
plt.show()
```

**히스토그램**

<br>

    hist()로 그릴 수 있음.

```python
plt.style.use('ggplot')
data1 = 100
data2 = 10
np.random.seed(1)
x = np.random.normal(data1, data2, 1000)

fig = plt.figure()
ax = fig.add_subplot(111)

ax.hist(x)
plt.show()
```

**Boxplot**

<br>

    boxplot()으로 그릴 수 있음.

```python
plt.style.use('ggplot')

x = [1, 2, 3, 3, 11, 20]

fig = plt.figure()
ax = fig.add_subplot(111)

ax.boxplot(x)
plt.show()
```

**Pie차트**

<br>

    pie()로 그릴 수 있음.

```python
plt.style.use('ggplot')

labels = ['english', 'math', 'korean']
sizes = [45, 30, 25]

fig = plt.figure(figsize=(3,3)) # 도화지 크기 미리 지정.
ax = fig.add_subplot(111)

ax.pie(sizes, labels = labels)
plt.show()
```
