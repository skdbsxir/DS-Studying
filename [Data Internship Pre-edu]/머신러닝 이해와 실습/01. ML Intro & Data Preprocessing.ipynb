## 회귀 분석

    입력 데이터를 바탕으로 원하는 타겟 변수의 미래 결과를 예측하는 작업.
    

- 통계학의 지식 많이 활용
- 회귀분석/다변량통계학/시계열분석 등을 주로 활용
- 예시?
    - 금융 분야 : 주식 시장 예측, 환율 예측, 상품 risk 예측 등...
    - CRM : 백화점 수요 예측, 가격 견적, 광고 입찰, 고객의 쿠폰 반응 예측 등...

<br>

## 분류

    입력 데이터를 바탕으로 개별 데이터의 class를 예측하거나 계급을 결정하는 작업.
    
- 예시? 스팸메일 필터링, 신용카드 사기탐지, VIP고객 여부, 제조 결함 탐지 등...

<br>

## 추천

    고객이 선호하는 상품 혹은 그 대안을 예측하는 작업.
    
- 예시? 홈쇼핑의 상품추천, 넷플릭스 영화추천 등...

<br>

## 대체

    누락된 입력 데이터의 값을 보강하는 작업.
    
  
- 예시? 손상된 고객 데이터 보충, 인구조사 자료 보충, 환자 의료 데이터 기록 보강 등...

# ML Workflow



<br>

- 모델이 안정화 될 때 까지 실 데이터를 통한 지속적인 Feedback이 필요.

<br>

**방법론 3가지**

    지도학습(target 데이터 존재)
    비지도학습(target 데이터 X. 숨겨진 구조 찾기)
    강화학습(target을 만들면서 훈련. 해답에서 멀어지면 벌점. 문제를 게임으로 간주해 적용)

### **일반적인 ML 구현 과정**

<br>


    데이터 가공 및 전처리 -> 훈련 데이터 추출 -> 모델 평가

<br>

- 전처리 단계에서 **결측치, 이상치 처리 작업** 필요.
    - Munging (가공), Wrangling (보강) 등의 작업. 왜곡 데이터는 모델 성능을 저하시키는 요인.
